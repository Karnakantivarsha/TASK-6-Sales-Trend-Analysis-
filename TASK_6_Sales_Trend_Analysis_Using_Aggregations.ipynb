```python
# üßÆ TASK 6: Sales Trend Analysis Using Aggregations
# Data Analyst Internship Project

# Step 1Ô∏è‚É£: Import Libraries
import sqlite3
import pandas as pd

# Step 2Ô∏è‚É£: Create a connection to SQLite (in-memory)
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Step 3Ô∏è‚É£: Create Table "online_sales"
cursor.execute('''
CREATE TABLE online_sales (
    order_id INTEGER,
    order_date TEXT,
    product_id INTEGER,
    amount REAL
)
''')

# Step 4Ô∏è‚É£: Insert Sample Data (you can add more rows if needed)
data = [
    (1, '2024-01-15', 101, 500),
    (2, '2024-01-28', 102, 1200),
    (3, '2024-02-02', 103, 700),
    (4, '2024-02-15', 104, 1500),
    (5, '2024-03-03', 105, 1800),
    (6, '2024-03-17', 106, 900),
    (7, '2024-04-01', 107, 2200),
    (8, '2024-04-25', 108, 800),
    (9, '2024-05-05', 109, 2500),
    (10, '2024-05-21', 110, 1000),
    (11, '2024-05-28', 111, 1500),
    (12, '2024-06-10', 112, 2700)
]
cursor.executemany("INSERT INTO online_sales VALUES (?, ?, ?, ?)", data)
conn.commit()

# Step 5Ô∏è‚É£: Display the Raw Data
print("‚úÖ Online Sales Dataset:")
df = pd.read_sql_query("SELECT * FROM online_sales", conn)
display(df)

# Step 6Ô∏è‚É£: Monthly Revenue + Order Volume
query_monthly = '''
SELECT
    STRFTIME('%Y', order_date) AS year,
    STRFTIME('%m', order_date) AS month,
    SUM(amount) AS total_revenue,
    COUNT(DISTINCT order_id) AS total_orders
FROM online_sales
GROUP BY year, month
ORDER BY year, month;
'''

print("üìä Monthly Revenue and Order Volume:")
df_monthly = pd.read_sql_query(query_monthly, conn)
display(df_monthly)

# Step 7Ô∏è‚É£: Top 3 Months by Total Revenue
query_top3 = '''
SELECT
    STRFTIME('%Y', order_date) AS year,
    STRFTIME('%m', order_date) AS month,
    SUM(amount) AS total_revenue
FROM online_sales
GROUP BY year, month
ORDER BY total_revenue DESC
LIMIT 3;
'''

print("üèÜ Top 3 Months by Revenue:")
df_top3 = pd.read_sql_query(query_top3, conn)
display(df_top3)

# Step 8Ô∏è‚É£: Close the Connection
conn.close()
```
